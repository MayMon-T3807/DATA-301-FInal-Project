```python
import pandas as pd
```

```python
movies = pd.read_csv('imdb_top_1000.csv')
movies.columns
```

```python
features = movies[["Series_Title", "Released_Year", "Genre", "IMDB_Rating", "Meta_score", "Director", "Star1", "Star2", "Star3", "Star4"]]
features
```

```python
features["Genre"].value_counts()
```

```python
genre_split = features['Genre'].str.split(', ', expand=True)

genre_split.columns = [f'genre_{i+1}' for i in genre_split.columns]

features = features.drop('Genre', axis=1).join(genre_split)

features.head()
```

```python
features.isna().sum()
```

```python
features.drop(["Meta_score"], axis=1, inplace=True)

features.head()
```

```python
features.fillna("None", inplace=True)
features.isna().sum()
```

```python
features.to_csv('imdb_top_1000_cleaned.csv', index=False)
```
